```python
import random
import matplotlib.pyplot as plt
import numpy as np
import math
import scipy.stats
```

```python
amount_of_samples = 1_000_000
data_x = scipy.stats.expon.rvs(size=amount_of_samples)
data_y = scipy.stats.expon.rvs(size=amount_of_samples)
plt.scatter(data_x,data_y,c='red',s=0.1)
plt.show()
```

```python
good_boys = 0
for i in range(amount_of_samples):
    good_boys += data_x[i] < data_y[i]

print(f"P(X<Y)= {good_boys/amount_of_samples}")
        
```

```python
good_boys = 0
for i in range(amount_of_samples):
    good_boys += data_x[i] + data_y[i] <= 1
print(f"P(X+Y<=1)= {good_boys/amount_of_samples}")
```

```python
sum = 0
for i in range(amount_of_samples):
    sum += data_x[i] + data_y[i]
print(f"E[X+Y]= {sum/amount_of_samples}")
```
