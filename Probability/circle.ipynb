```python
import random
import matplotlib.pyplot as plt
import numpy as np
import math

reps = 50_000
low = -1
high = 1
data_x = []
data_y = []
for i in range(reps):
    x_i = random.uniform(low,high)
    y_i = random.uniform(low,high)
    while x_i ** 2 + y_i ** 2 >= 1:
        x_i = random.uniform(low,high)
        y_i = random.uniform(low,high)
    data_x.append(x_i)
    data_y.append(y_i)

# plt.plot(data_x,data_y)
plt.scatter(data_x, data_y, c='blue', s=0.1)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

```python
sum = 0
for i in range(reps):
    sum += math.sqrt(data_x[i] ** 2 + data_y[i] ** 2)
mean = sum / reps
print(f"The mean is: {mean}")
```

```python
sum = 0
for i in range(reps):
    sum += (mean - math.sqrt(data_x[i] ** 2 + data_y[i] ** 2)) ** 2
variance = sum / reps
print(f"The variance is: {variance}")
print(f"1/18 = {1/18}")
print(f"The standard deviation is: {math.sqrt(variance)}")
```
